```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
base_dir = "Data"
filename_1 = '01-21 HVCH FULL EOB.csv'
filename_2 = '02-21 HVCH FULL EOB.csv'
filename_3 = 'EOB_Description_Selected.xlsx'
```

```python
df_1 = pd.read_csv(os.path.join(base_dir, filename_1), 
                   low_memory=False
                  )
df_1.head()
```

```python
df_2 = pd.read_csv(os.path.join(base_dir, filename_2),
                   low_memory=False
                  )
df_2.head()
```

```python
mapping_df = pd.read_excel(os.path.join(base_dir, filename_3))
mapping_df.head()
```

```python
df_1['Remit Med Rec Num'].isnull().sum()
```

```python
# Filtering the Remit Med Rec Num starting with HV
data_1 = df_1[df_1['Remit Med Rec Num'].astype(str).str.startswith('HV')]
print(data_1.shape)
data_1.head()
```

```python
# Filtering the Remit Med Rec Num starting with HV
data_2 = df_2[df_2['Remit Med Rec Num'].astype(str).str.startswith('HV')]
print(data_2.shape)
data_2.head()
```

```python
# Appending both the files
merged_df = pd.concat([data_1, data_2], axis=0, ignore_index=True)
print("shape after merge: ", merged_df.shape)
```

```python
# Taking only selected columns
cols_req = mapping_df['Variables'].tolist()
merged_df = merged_df[cols_req].copy()
merged_df.head()
```
